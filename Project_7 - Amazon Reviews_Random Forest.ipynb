# AMAZON ALEXA REVIEW RATINGS CLASSIFICATION 


# STEP #1: PROBLEM STATEMENT

- Dataset consists of 3000 Amazon customer reviews, star ratings, date of review, variant and feedback of various amazon Alexa products like Alexa Echo, Echo dots.
- The objective is to discover insights into consumer reviews and perfrom sentiment analysis on the data.
- Dataset: www.kaggle.com/sid321axn/amazon-alexa-reviews

# STEP#4: DATA CLEANING/FEATURE ENGINEERING

```python
dataset
```

```python
dataset = dataset.drop(columns = ['rating', 'date'])
```

```python
dataset
```

```python
variation_dummies = pd.get_dummies(dataset['variation'], drop_first = True)
```

```python
variation_dummies
```

```python
dataset = dataset.drop(columns = ['variation'])
```

```python
dataset
```

```python
dataset = pd.concat([dataset, variation_dummies], axis = 1)
```

```python
dataset
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
alexa_vectorizer = vectorizer.fit_transform(dataset['verified_reviews'])
```

```python
alexa_vectorizer.shape
```

```python
print(vectorizer.get_feature_names())
```

```python
print(alexa_vectorizer.toarray())
```

```python
dataset = dataset.drop(columns = ['verified_reviews'])
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
encoded_reviews = pd.DataFrame(alexa_vectorizer.toarray())
```

```python
dataset = pd.concat([dataset, encoded_reviews], axis=1)
```

```python
dataset.head()
```

```python
X = dataset.drop(columns = ['feedback'])
```

```python
X
```

```python
X.shape
```

```python
y = dataset['feedback']
```

```python
y
```

```python
y.shape
```

# STEP #5: MODEL TRAINING 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
random_forest_classifier = RandomForestClassifier(n_estimators=100, criterion='entropy')
random_forest_classifier.fit(X_train, y_train)
```

# STEP #6: EVALUATING THE MODEL

```python
y_pred_train = random_forest_classifier.predict(X_train)
```

```python
y_pred_train
```

```python
y_train
```

```python
cm = confusion_matrix(y_train, y_pred_train)
```

```python
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_train, y_pred_train))
```

```python
y_pred_test = random_forest_classifier.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred_test)
```

```python
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test, y_pred_test))
```

# STEP #7 IMPROVE THE MODEL

```python
dataset = pd.read_csv("amazon_alexa.tsv", sep = "\t")
```

```python
dataset = pd.concat([dataset, pd.DataFrame(alexa_vectorizer.toarray())], axis = 1)
```

```python
dataset
```

```python
X = dataset.drop(columns = ['rating', 'date', 'variation', 'verified_reviews', 'feedback'])
```

```python
y = dataset['feedback']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 5)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
random_forest_classifier = RandomForestClassifier(n_estimators=100, criterion='entropy')
random_forest_classifier.fit(X_train, y_train)
```

```python
y_pred = random_forest_classifier.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 5)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
random_forest_classifier = RandomForestClassifier(n_estimators=150, criterion='entropy')
random_forest_classifier.fit(X_train, y_train)
```

```python
y_pred = random_forest_classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot = True)
print(classification_report(y_test, y_pred))
```
