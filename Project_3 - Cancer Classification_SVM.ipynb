# BREAST CANCER CLASSIFICATION USING SUPPORT VECTOR MACHINES


# STEP #1: PROBLEM STATEMENT


- Predicting if the cancer diagnosis is benign or malignant based on several observations/features 
- 30 features are used, examples:
        - radius (mean of distances from center to points on the perimeter)
        - texture (standard deviation of gray-scale values)
        - perimeter
        - area
        - smoothness (local variation in radius lengths)
        - compactness (perimeter^2 / area - 1.0)
        - concavity (severity of concave portions of the contour)
        - concave points (number of concave portions of the contour)
        - symmetry 
        - fractal dimension ("coastline approximation" - 1)

- Datasets are linearly separable using all 30 input features
- Number of Instances: 569
- Class Distribution: 212 Malignant, 357 Benign
- Target class:
         - Malignant
         - Benign


https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+(Diagnostic)

# STEP #4: MODEL TRAINING (FINDING A PROBLEM SOLUTION)

```python
X = cancer_df.drop('target', axis = 1)
```

```python
y = cancer_df['target']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
```

```python
cancer_svc = SVC()
```

```python
cancer_svc.fit(X_train, y_train)
```

# STEP #5: EVALUATING THE MODEL

```python
y_pred = cancer_svc.predict(X_test)
```

```python
y_pred
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test, y_pred))
```

# STEP #6: IMPROVING THE MODEL

```python
X_train_min = X_train.min()
```

```python
X_train_min
```

```python
X_train_range = (X_train - X_train_min).max()
```

```python
X_train_range
```

```python
X_train_scaled = (X_train - X_train_min) / X_train_range
```

```python
X_train_scaled
```

```python
sns.scatterplot(x = X_train['mean area'], y = X_train['mean smoothness'], hue = y_train)
```

```python
sns.scatterplot(x = X_train_scaled['mean area'], y = X_train_scaled['mean smoothness'], hue = y_train)
```

```python
X_test_min = X_test.min()
```

```python
X_test_min
```

```python
X_test_range = (X_test - X_test_min).max()
```

```python
X_test_scaled = (X_test - X_test_min) / X_test_range
```

```python
X_test_scaled
```

```python
sns.scatterplot(x = X_test['mean radius'], y = X_test['mean smoothness'], hue = y_test)
```

```python
sns.scatterplot(x = X_test_scaled['mean radius'], y = X_test_scaled['mean smoothness'], hue = y_test)
```

```python
cancer_svc.fit(X_train_scaled, y_train)
```

```python
y_pred = cancer_svc.predict(X_test_scaled)
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test, y_pred))
```

# IMPROVING THE MODEL - PART 2

```python
param_grid = {'C': [0.1, 1, 10, 100], 'gamma': [1, 0.1, 0.01, 0.001], 'kernel': ['rbf']}
```
